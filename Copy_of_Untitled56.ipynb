```python
import gradio as gr
import google.generativeai as genai
import requests
import json
import os
from bs4 import BeautifulSoup
from urllib.parse import urljoin
from dateutil.parser import parse
from functools import reduce
from itertools import chain

# Gemini API Configuration

GOOGLE_API_KEY = os.environ.get("AIzaSyAcJUGcSNV6Fxg9jVAVZQbqBYpUn8zFoXk")
genai.configure(api_key="AIzaSyAcJUGcSNV6Fxg9jVAVZQbqBYpUn8zFoXk")
model = genai.GenerativeModel('gemini-1.5-pro')
chat = model.start_chat(history=[])

def fetchIt(url: str) -> str:
    """Fetch content of target webpage"""
    try:
        resp = requests.get(url)
        return resp.content if resp.status_code == 200 else None
    except Exception:
        return None

def extractIt(dataObject, username: str, targetClass: str = 'col-10 col-lg-9 d-inline-block'):
    """Extract repository information from GitHub profile"""
    def __extractFromThisPage__():
        def __findURL__(element) -> str:
            try:
                return urljoin('https://github.com', element.findChild('div', {'class': 'd-inline-block mb-1'}).h3.a.get('href'))
            except Exception:
                return ''

        def __findDescription__(element) -> str:
            try:
                return element.findChild('p', {'class': 'col-9 d-inline-block text-gray mb-2 pr-4'}).text.strip()
            except Exception:
                return ''

        def __findLanguage__(element) -> str:
            try:
                return element.findChild('div', {'class': 'f6 text-gray mt-2'}).findChild('span', {'itemprop': 'programmingLanguage'}).text.strip()
            except Exception:
                return ''

        def __findLanguageColor__(element) -> str:
            try:
                return element.findChild('div', {'class': 'f6 text-gray mt-2'}).findChild('span', {'class': 'ml-0 mr-3'}).findChild('span', {'class': 'repo-language-color'}).get('style').split(':')[1].strip()
            except Exception:
                return ''

        def __findForkCount__(element) -> int:
            try:
                return reduce(lambda acc, cur: acc + int(cur.text.strip()),
                               filter(lambda e: e.get('href').endswith('members'),
                                      element.findChild('div', {'class': 'f6 text-gray mt-2'}).findAll('a', {'class': 'muted-link mr-3'})), 0)
            except Exception:
                return 0

        def __findStargazerCount__(element) -> int:
            try:
                return reduce(lambda acc, cur: acc + int(cur.text.strip()),
                               filter(lambda e: e.get('href').endswith('stargazers'),
                                      element.findChild('div', {'class': 'f6 text-gray mt-2'}).findAll('a', {'class': 'muted-link mr-3'})), 0)
            except Exception:
                return 0

        def __findLastUpdated__(element) -> float:
            try:
                return parse(element.findChild('div', {'class': 'f6 text-gray mt-2'}).findChild('relative-time').get('datetime')).timestamp()
            except Exception:
                return 0.0

        def __findLicense(element) -> str:
            try:
                return reduce(lambda acc, cur: acc + cur.text.strip(),
                               filter(lambda e: ''.join(e.get('class')) == 'mr-3',
                                      element.findChild('div', {'class': 'f6 text-gray mt-2'}).findAll('span', {'class': 'mr-3'})), '')
            except Exception:
                return ''

        def __checkIfForked__(element) -> str:
            try:
                return urljoin('https://github.com',
                                element.findChild('div', {'class': 'd-inline-block mb-1'}).findChild('span', {'class': 'f6 text-gray mb-1'}).findChild('a', {'class': 'muted-link'}).get('href'))
            except Exception:
                return ''

        return map(lambda e: (e.findChild('div', {'class': 'd-inline-block mb-1'}).h3.a.text.strip(), {
            'url': __findURL__(e),
            'description': __findDescription__(e),
            'lang': __findLanguage__(e),
            'langColor': __findLanguageColor__(e),
            'fork': __findForkCount__(e),
            'star': __findStargazerCount__(e),
            'updated': __findLastUpdated__(e),
            'license': __findLicense(e),
            'forkedFrom': __checkIfForked__(e)}),
            dataObject.findAll('div', {'class': targetClass}))

    try:
        paginatedObject = dataObject.findChild('div', {'class': 'paginate-container'})
        if paginatedObject:
            linksToFollow = paginatedObject.div.findAll('a')
            if len(linksToFollow) == 2:
                return chain(__extractFromThisPage__(),
                             extractIt(BeautifulSoup(fetchIt(linksToFollow[1].get('href')), features='html.parser'),
                                       username=username))
            else:
                if linksToFollow[0].text.strip().lower() == 'next':
                    return chain(__extractFromThisPage__(),
                                 extractIt(BeautifulSoup(fetchIt(linksToFollow[0].get('href')), features='html.parser'),
                                           username=username))
                else:
                    return __extractFromThisPage__()
        else:
            return __extractFromThisPage__()
    except Exception:
        return None

def extract_github_repositories(username):
    """Extract repositories for a given GitHub username"""
    try:
        # Fetch repositories page
        url = f'https://github.com/{username}?tab=repositories'
        content = fetchIt(url)

        if not content:
            return "Failed to fetch repositories"

        # Parse the content
        soup = BeautifulSoup(content, features='html.parser')

        # Extract repositories
        repos = dict(extractIt(soup, username))

        return repos
    except Exception as e:
        return f"Error: {str(e)}"

def analyze_repositories_with_gemini(repos, analysis_type):
    """Analyze repositories using Gemini"""
    system_prompt = f"""
    You are a GitHub Repository Analysis Assistant specializing in {analysis_type}.
    Analyze the provided repository data and provide comprehensive insights.
    Focus on providing actionable recommendations and detailed analysis.
    """

    # Prepare repository data for analysis
    repo_data_str = json.dumps(repos, indent=2)

    # Send to Gemini for analysis
    try:
        chat.history = [{"parts": [{"text": system_prompt}], "role": "user"}]
        response = chat.send_message(f"Repository Data:\n{repo_data_str}\n\nProvide a detailed {analysis_type} analysis.")
        response.resolve()
        return response.text
    except Exception as e:
        return f"Analysis error: {str(e)}"

# Gradio Interface
css = """
.gradio-container {
    font-family: 'GitHub', sans-serif;
    background-color: #0d1117;
    color: #c9d1d9;
}
.output-text {
    background-color: #161b22;
    color: #c9d1d9;
    border: 1px solid #30363d;
    border-radius: 6px;
    padding: 15px;
    white-space: pre-wrap;
    max-height: 500px;
    overflow-y: auto;
}
"""

with gr.Blocks(css=css) as demo:
    gr.Markdown("# üïµÔ∏è GitHub Repository Deep Analyzer")

    with gr.Row():
        username_input = gr.Textbox(label="GitHub Username", placeholder="Enter GitHub username")
        analysis_type = gr.Dropdown(
            label="Analysis Type",
            choices=[
                "Comprehensive Overview",
                "Language Distribution",
                "Repository Performance",
                "Community Engagement",
                "Technical Diversity",
                "Open Source Contribution"
            ],
            value="Comprehensive Overview"
        )

    analyze_btn = gr.Button("üîç Analyze Repositories")

    # Repository Data Output
    repo_output = gr.Textbox(label="Repository Analysis", elem_classes="output-text")

    # Repository Details Output
    details_output = gr.JSON(label="Repository Details")

    def process_analysis(username, analysis_type):
        # Extract repositories
        repos = extract_github_repositories(username)

        if isinstance(repos, str) and repos.startswith("Error"):
            return repos, {}

        # Analyze repositories
        analysis = analyze_repositories_with_gemini(repos, analysis_type)

        return analysis, repos

    analyze_btn.click(
        process_analysis,
        inputs=[username_input, analysis_type],
        outputs=[repo_output, details_output]
    )

    gr.Markdown("""
    ## üöÄ How to Use
    1. Enter a GitHub username
    2. Choose an analysis type
    3. Click 'Analyze Repositories'
    4. Get detailed insights about the user's GitHub repositories

    Powered by GitHub Scraping and Gemini AI Analysis ü§ñ
    """)

# Launch the app
if __name__ == "__main__":
    demo.launch(debug=True)
```

```python
import gradio as gr
import google.generativeai as genai
from github import Github
from neo4j import GraphDatabase
import requests
from bs4 import BeautifulSoup
from urllib.parse import urlparse, urljoin
from dateutil.parser import parse
from functools import reduce
from itertools import chain
import os
import json
import base64
import asyncio
import traceback

class GitHubInsightBot:
    def __init__(self, github_token, neo4j_uri, neo4j_user, neo4j_password, gemini_api_key):
        # GitHub and Neo4j initialization
        self.github = Github(github_token)
        self.driver = GraphDatabase.driver(neo4j_uri, auth=(neo4j_user, neo4j_password))

        # Gemini AI initialization
        genai.configure(api_key=gemini_api_key)
        self.model = genai.GenerativeModel('gemini-1.5-pro')
        self.chat = self.model.start_chat(history=[])

    def extract_repo_info(self, github_url):
        """Extract owner and repo name from GitHub URL"""
        parsed = urlparse(github_url)
        path_parts = parsed.path.strip('/').split('/')
        if len(path_parts) >= 2:
            return f"{path_parts[0]}/{path_parts[1]}"
        raise ValueError("Invalid GitHub URL format")

    async def analyze_repository(self, repo):
        """Comprehensive repository analysis and Neo4j storage"""
        with self.driver.session() as session:
            try:
                # Create or update repository node
                session.run("""
                    MERGE (r:Repository {full_name: $repo_name})
                    SET r.description = $description,
                        r.language = $language,
                        r.stars = $stars,
                        r.forks = $forks,
                        r.last_analyzed = datetime()
                """, repo_name=repo.full_name,
                   description=repo.description or "",
                   language=repo.language or "Unknown",
                   stars=repo.stargazers_count,
                   forks=repo.forks_count)

                # Analyze repository files
                contents = repo.get_contents("")
                files_to_analyze = []

                while contents:
                    file_content = contents.pop(0)
                    if file_content.type == "dir":
                        contents.extend(repo.get_contents(file_content.path))
                    else:
                        if file_content.name.endswith(('.py', '.js', '.java', '.cpp', '.go')):
                            files_to_analyze.append(file_content)

                # Deep analysis of files
                for file in files_to_analyze[:5]:  # Limit to 5 files
                    try:
                        content = base64.b64decode(file.content).decode('utf-8', errors='ignore')
                        await self.analyze_file(file.name, content, repo.full_name, session)
                    except Exception as file_error:
                        print(f"Error analyzing file {file.name}: {file_error}")

                # Extract contributors
                contributors = repo.get_contributors()
                for contributor in contributors[:10]:  # Limit to top 10
                    session.run("""
                        MERGE (u:User {login: $login})
                        SET u.name = $name,
                            u.contributions = $contributions
                        WITH u
                        MATCH (r:Repository {full_name: $repo_name})
                        MERGE (u)-[:CONTRIBUTED_TO]->(r)
                    """, login=contributor.login,
                       name=contributor.name or contributor.login,
                       contributions=contributor.contributions,
                       repo_name=repo.full_name)

                return "Repository analysis completed successfully!"

            except Exception as e:
                error_msg = f"Error analyzing repository {repo.full_name}: {e}"
                print(error_msg)
                traceback.print_exc()
                return error_msg

    async def analyze_file(self, filename, content, repo_name, session):
        """Detailed file analysis using Gemini"""
        prompt = f"""
        Analyze this code file and provide deep insights:
        1. Code architecture and patterns
        2. Potential security vulnerabilities
        3. Code complexity and maintainability
        4. Best practices and design patterns

        File: {filename}
        Content (first 2000 chars): {content[:2000]}

        Provide analysis in structured JSON format with keys:
        - patterns: List of code patterns
        - vulnerabilities: List of potential security issues
        - complexity: Numerical complexity score
        - recommendations: List of improvement suggestions
        """

        try:
            response = await self.model.generate_content(prompt)
            insights = json.loads(response.text)

            # Store file insights
            session.run("""
                MERGE (f:File {name: $filename, repo: $repo_name})
                SET f.complexity = $complexity,
                    f.patterns = $patterns,
                    f.vulnerabilities = $vulnerabilities,
                    f.recommendations = $recommendations
                WITH f
                MATCH (r:Repository {full_name: $repo_name})
                MERGE (r)-[:CONTAINS_FILE]->(f)
            """, filename=filename,
               repo_name=repo_name,
               complexity=insights.get('complexity', 0),
               patterns=json.dumps(insights.get('patterns', [])),
               vulnerabilities=json.dumps(insights.get('vulnerabilities', [])),
               recommendations=json.dumps(insights.get('recommendations', [])))

        except Exception as e:
            print(f"Error analyzing file {filename}: {e}")

    async def query_repository_insights(self, query, repo_name):
        """Natural language query processing"""
        # Generate a contextual query using Gemini
        context_prompt = f"""
        Convert this natural language query into a precise Neo4j Cypher query:
        Repository: {repo_name}
        Query: {query}

        Provide a Cypher query that extracts relevant information.
        Ensure the query is safe and doesn't modify the database.
        """

        try:
            cypher_response = await self.model.generate_content(context_prompt)
            cypher_query = cypher_response.text.strip()

            with self.driver.session() as session:
                results = session.run(cypher_query, repo_name=repo_name)
                return [dict(record) for record in results]

        except Exception as e:
            return [{"error": str(e)}]

def create_gradio_interface():
    with gr.Blocks() as demo:
        gr.Markdown("# üîç GitHub Repository Insight Bot")

        # Credential Input Section
        with gr.Accordion("üîê Credentials", open=False):
            with gr.Row():
                github_token = gr.Textbox(label="GitHub Personal Access Token", type="password")
                gemini_api_key = gr.Textbox(label="Gemini API Key", type="password")

            with gr.Row():
                neo4j_uri = gr.Textbox(label="Neo4j URI", placeholder="neo4j+s://your-neo4j-instance.databases.neo4j.io")
                neo4j_user = gr.Textbox(label="Neo4j Username", value="neo4j")
                neo4j_password = gr.Textbox(label="Neo4j Password", type="password")

        # Main Analysis Section
        with gr.Row():
            github_url = gr.Textbox(label="GitHub Repository URL", placeholder="https://github.com/owner/repo")
            analysis_type = gr.Dropdown(
                label="Analysis Focus",
                choices=[
                    "Technical Overview",
                    "Security Analysis",
                    "Code Quality",
                    "Contributor Insights",
                    "Architectural Overview"
                ],
                value="Technical Overview"
            )

        # Action Buttons
        with gr.Row():
            analyze_btn = gr.Button("üöÄ Analyze Repository")
            query_input = gr.Textbox(label="Ask about the Repository")
            query_btn = gr.Button("üîç Query Insights")

        # Output Components
        repo_insights = gr.Textbox(label="Repository Insights")
        query_result = gr.JSON(label="Query Results")

        # Async Analysis Function
        async def analyze_repo(github_token, gemini_api_key, neo4j_uri,
                               neo4j_user, neo4j_password,
                               github_url, analysis_type):
            try:
                # Validate inputs
                if not all([github_token, gemini_api_key, neo4j_uri,
                            neo4j_user, neo4j_password, github_url]):
                    return "Please fill in all credential and repository fields."

                # Initialize Bot
                bot = GitHubInsightBot(
                    github_token=github_token,
                    neo4j_uri=neo4j_uri,
                    neo4j_user=neo4j_user,
                    neo4j_password=neo4j_password,
                    gemini_api_key=gemini_api_key
                )

                # Extract and analyze repo
                repo_name = bot.extract_repo_info(github_url)
                repo = bot.github.get_repo(repo_name)

                # Perform repository analysis
                analysis_result = await bot.analyze_repository(repo)

                # Generate detailed insights using Gemini
                prompt = f"""
                Provide a comprehensive {analysis_type} for repository: {repo_name}
                Include:
                - Key technical insights
                - Architectural overview
                - Potential improvements
                - Notable patterns and practices
                """

                response = await bot.model.generate_content(prompt)
                return response.text

            except Exception as e:
                error_trace = traceback.format_exc()
                return f"Analysis Error: {str(e)}\n\nDetailed Trace:\n{error_trace}"

        # Async Query Function
        async def query_repo(github_token, gemini_api_key, neo4j_uri,
                             neo4j_user, neo4j_password,
                             github_url, query):
            try:
                # Validate inputs
                if not all([github_token, gemini_api_key, neo4j_uri,
                            neo4j_user, neo4j_password, github_url]):
                    return [{"error": "Please fill in all credential and repository fields."}]

                # Initialize Bot
                bot = GitHubInsightBot(
                    github_token=github_token,
                    neo4j_uri=neo4j_uri,
                    neo4j_user=neo4j_user,
                    neo4j_password=neo4j_password,
                    gemini_api_key=gemini_api_key
                )

                # Extract repo name and query insights
                repo_name = bot.extract_repo_info(github_url)
                results = await bot.query_repository_insights(query, repo_name)
                return results

            except Exception as e:
                error_trace = traceback.format_exc()
                return [{"error": f"Query Error: {str(e)}", "trace": error_trace}]

        # Event Handlers
        analyze_btn.click(
            analyze_repo,
            inputs=[
                github_token, gemini_api_key, neo4j_uri,
                neo4j_user, neo4j_password,
                github_url, analysis_type
            ],
            outputs=repo_insights
        )

        query_btn.click(
            query_repo,
            inputs=[
                github_token, gemini_api_key, neo4j_uri,
                neo4j_user, neo4j_password,
                github_url, query_input
            ],
            outputs=query_result
        )

        # Usage Guide
        gr.Markdown("""
        ## ü§ñ GitHub Repository Insight Bot Guide

        ### Prerequisites
        - GitHub Personal Access Token
          - Create at: https://github.com/settings/tokens
          - Needs `repo` scope
        - Gemini API Key
          - Get from Google AI Studio
        - Neo4j Database
          - Can use Neo4j Aura (free tier)

        ### How to Use
        1. Open the Credentials accordion
        2. Enter your API keys and database credentials
        3. Provide a GitHub Repository URL
        4. Choose an Analysis Type
        5. Click "Analyze Repository"
        6. Ask specific questions about the repository

        ### Example Queries
        - "What are the main programming languages?"
        - "List top contributors"
        - "Show security vulnerabilities"
        - "Describe code architecture"
        """)

    return demo

if __name__ == "__main__":
    demo = create_gradio_interface()
    demo.launch(debug=True)
```
